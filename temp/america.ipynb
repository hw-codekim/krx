```python
# 전 종목 파일 생성

today = datetime.today().strftime("%Y%m%d") #오늘
for ix, row in sp500.iterrows():

    sym, name = row['Symbol'], row['Name']
    csv_fn = os.path.join(folder, f'{sym}.csv')
    if os.path.exists(csv_fn):
        continue
      
    try:
      df = fdr.DataReader(sym, '2024-01-02', today)
      df['Symbol'] = sym
      
      df.to_csv(csv_fn, quoting=csv.QUOTE_MINIMAL)
    except ValueError as e:
      print(e)
    except IndexError as e:
      print(e)
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.keys import Keys
import time
import pandas as pd

# 웹드라이버 설정 (ChromeDriver 경로 설정 필요)
driver = webdriver.Chrome()

# URL로 접속
url = "http://data.krx.co.kr/contents/MDC/MDI/mdiLoader/index.cmd?menuId=MDC0201020302"
driver.get(url)

# 1. 첫 번째 XPath 클릭 (항목 선택)
first_xpath = '//*[@id="MDCSTAT023_FORM"]/div[1]/div/table/tbody/tr[1]/td/label[2]'
WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, first_xpath))).click()

# 2. 두 번째 XPath 클릭 (항목 선택)
second_xpath = '//*[@id="MDCSTAT023_FORM"]/div[1]/div/table/tbody/tr[3]/td[1]/div/div/button[2]'
WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, second_xpath))).click()

# 3. 검색창에 값 입력 (예시: '삼성전자' 코드 입력)
search_box_xpath = '//*[@id="MDCSTAT023_FORM"]/div[1]/div/table/tbody/tr[2]/td/div/div/p/div'

search_code = '005930'  # 검색할 종목 코드
search_box = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, search_box_xpath)))

# 3-1. 체크박스 클릭 (추가된 부분)
checkbox_xpath = '//*[@id="MDCSTAT023_FORM"]/div[1]/div/table/tbody/tr[3]/td[2]/div/label'
checkbox = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, checkbox_xpath)))
checkbox.click()

# 4. 조회 버튼 클릭
search_button_xpath = '//*[@id="jsSearchButton"]'
WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, search_button_xpath))).click()

# 조회 후 표가 로드될 때까지 대기
time.sleep(5)  # 혹은 명시적으로 해당 요소가 로드될 때까지 기다릴 수 있음

# 5. 표 데이터 가져오기 (XPath를 이용)


# 1. 대상 요소를 찾기
element_xpath = '//*[@id="jsMdiContent"]/div[2]/div[5]/div[1]/div[2]'
element = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, element_xpath)))

table_rows_xpath = '//table[@class="CI-GRID-BODY-TABLE"]/tbody/tr[@class="CI-GRID-EVEN"]'
# 2. 오른쪽 끝까지 스크롤
driver.execute_script("arguments[0].scrollLeft = arguments[0].scrollWidth", element)
rows = WebDriverWait(driver, 10).until(EC.presence_of_all_elements_located((By.XPATH, table_rows_xpath)))

# 데이터프레임으로 변환
data = []
for row in rows:

    # 각 행에서 모든 열(<td>)을 찾음
    cols = row.find_elements(By.TAG_NAME, 'td')
    
    # 각 열의 텍스트를 추출하여 리스트로 만듦
    cols_text = [col.text for col in cols]
    
    # 열의 개수가 일치하지 않으면, 공백을 추가하여 열 개수를 맞춤
    if len(cols_text) < 13:
        cols_text.extend([''] * (13 - len(cols_text)))
    
    # 열 중 하나라도 값이 있으면 데이터를 추가
    if any(cols_text):
        data.append(cols_text)

# Pandas DataFrame으로 변환
df = pd.DataFrame(data, columns=['일자', '금융투자', '보험', '투신', '사모', '은행', '기타금융', '연기금 등', '기타법인', '개인', '외국인', '기타외국인', '전체'])


# 결과 출력
print(df)

# 크롬 드라이버 종료
driver.quit()
```

```python
# FinanceDataReader 라이브러리 불러오기 (Bokeh는 따로 안불러와도 됨)
import FinanceDataReader as fdr

# 삼성전자 주가 데이터 가져오기 (2020년 1월 1일부터 현재까지)
def corp_price(code):
    df = fdr.DataReader(code, '2024-01-01')
    df['c5'] = df['Close'].shift(5)
    df['c20'] = df['Close'].shift(20)
    df['c60'] = df['Close'].shift(60)
    df['c120'] = df['Close'].shift(120)
    df['v5'] = df['Volume'].shift(5)
    df['v20'] = df['Volume'].shift(20)
    df['v60'] = df['Volume'].shift(60)
    df['v120'] = df['Volume'].shift(120)
    
    # 5일선 > 120일
    df['score'] = 0
    
    if df.iloc[-1]['c5'] > df.iloc[-1]['c120']:
        df['score'] += 1
    
    if df.iloc[-1]['c5'] > df.iloc[-1]['c20']:
        df['score'] += 1

    if df.iloc[-1]['c5'] > df.iloc[-1]['c60']:
        df['score'] += 1

    if df.iloc[-1]['v5'] > df.iloc[-1]['v120']:
        df['score'] += 1
    
    if df.iloc[-1]['v5'] > df.iloc[-1]['v20']:
        df['score'] += 1

    if df.iloc[-1]['v5'] > df.iloc[-1]['v60']:
        df['score'] += 1

        
    return df

co = corp_price('077360')
co.tail(50)
```
